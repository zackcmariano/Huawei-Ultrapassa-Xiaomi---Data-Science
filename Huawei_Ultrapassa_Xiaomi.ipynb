##Grfico semestral de acessos da Web Mundial:

```
#IMPORTING LIBRARIES
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib import pyplot as pls
import numpy as np
%matplotlib inline

plt.style.use("ggplot")

acess_web = pd.read_csv('https://raw.githubusercontent.com/zackcmariano/Huawei-enconsta-em-Xiaomi---Data-Science/main/vendor-ww-monthly-202008-202101.csv')
```

```
#FIRST INFORMATION
acess_web.head(5)
```

```
#PLOTING THE GRAPHIC
data = acess_web.Date
year = ['Ago/20','Set/20','Out/20','Nov/20','Dez/20','Jan/21']
x = np.arange(len(data))

fig,ax = plt.subplots(figsize=(18,6))

Samsung = ax.bar(x - 0.20, acess_web.Samsung, width = 0.20, color = '#23807e')
Apple = ax.bar(x, acess_web.Apple, width = 0.20, color = '#6ae1e2')
Huawei = ax.bar(x + 0.20, acess_web.Huawei, width = 0.20, color = '#2e95b4')
Xiaomi = ax.bar(x + 0.40, acess_web.Xiaomi, width= 0.20, color = '#42c5ba')

plt.xticks(fontsize=14)
ax.set_title('LIDERANA MOBILE NA WEB', fontsize=20, color='#1f6376')
ax.set_ylabel('Porcentagem %', fontsize=15, color='#25424a')

ax.set_ylim([0,100])
ax.set_xticks(x)
ax.set_xticklabels(year)

plt.legend(labels=['Samsung', 'Apple', 'Huawei', 'Xiaomi'], fontsize=14);

def autolabel(system):
  for i in system:
    percentage = i.get_height()
    ax.annotate('{}'.format(percentage),
                xy = (i.get_x()+i.get_width()/2,percentage),
                xytext = (0,5),
                textcoords = 'offset points',
                ha = 'center', fontsize=12)    

autolabel(Samsung)
autolabel(Apple)
autolabel(Huawei)
autolabel(Xiaomi)

#plt.savefig('BrowsersBr.png',dpi = 600)   / TO SAVE PNG

plt.show()
```
